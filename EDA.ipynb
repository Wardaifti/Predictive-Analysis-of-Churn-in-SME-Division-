# **Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:

        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot for churn distribution
if 'churn' in client_df.columns:
    client_df['churn'].value_counts().plot(kind='bar', figsize=(6, 4))
    plt.title('Churn Distribution')
    plt.ylabel('Count')
    plt.xlabel('Churn Status')
    plt.show()
else:
    print("Column 'churn' not found in client_df")
```

```python

# Boxplot for Usage by Churn
# Assuming a column for usage exists (like 'cons_12m' or 'imp_cons')
if 'churn' in client_df.columns and 'imp_cons' in client_df.columns:
    sns.boxplot(x='churn', y='imp_cons', data=client_df)
    plt.title('Usage by Churn Status')
    plt.xlabel('Churn Status')
    plt.ylabel('Usage (imp_cons)')
    plt.show()
else:
    print("Columns 'churn' or 'imp_cons' not found in client_df")
```

```python
# Bar plot for fixed prices (off-peak, peak, mid-peak)
fixed_price_cols = ['price_off_peak_fix', 'price_peak_fix', 'price_mid_peak_fix']
if all(col in price_df.columns for col in fixed_price_cols):
    price_df[fixed_price_cols].mean().plot(kind='bar', figsize=(8, 5), color=['blue', 'orange', 'green'])
    plt.title('Average Fixed Prices')
    plt.ylabel('Price')
    plt.xlabel('Fixed Price Categories')
    plt.xticks(rotation=0)
    plt.show()
else:
    print("Some fixed price columns are missing in price_df")

```

```python

# Boxplot for Variable Prices
variable_price_cols = ['price_off_peak_var', 'price_peak_var', 'price_mid_peak_var']
if all(col in price_df.columns for col in variable_price_cols):
    melted_df = price_df.melt(value_vars=variable_price_cols, var_name='Variable Price Category', value_name='Price')
    sns.boxplot(x='Variable Price Category', y='Price', data=melted_df)
    plt.title('Distribution of Variable Prices')
    plt.xticks(rotation=15)
    plt.show()
else:
    print("Some variable price columns are missing in price_df")
```

```python
# Convert date columns to datetime
client_df['date_activ'] = pd.to_datetime(client_df['date_activ'])
client_df['year_activ'] = client_df['date_activ'].dt.year

# Plot number of sign-ups by year
client_df['year_activ'].value_counts().sort_index().plot(kind='bar', figsize=(8, 5))
plt.title("Customer Sign-Ups by Year")
plt.xlabel("Year")
plt.ylabel("Number of Sign-Ups")
plt.show()
```
